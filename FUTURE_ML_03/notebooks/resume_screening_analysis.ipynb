# Resume / Candidate Screening System

                This notebook walks through a simple NLP-based resume screening workflow.
                It compares resumes against a job description, extracts key skills, scores candidates,
                and produces a ranked list for HR or recruitment teams.


## Problem Statement

                Recruiters often review many resumes for one role. This project helps by:
                - cleaning and standardizing resume text
                - extracting important skills
                - measuring resume similarity with the job description
                - ranking candidates based on a combined score

                ## Sample Job Description

                We are hiring a Machine Learning Engineer for an HR-tech resume screening platform.

The role requires strong Python and SQL skills, hands-on machine learning knowledge, and practical experience with NLP pipelines. Candidates should understand TF-IDF, cosine similarity, data preprocessing, feature engineering, and model evaluation. Experience with spaCy, scikit-learn, pandas, and Excel is a plus.

The ideal candidate can build ranking systems, extract skills from resumes, communicate results clearly, and work with recruiters and business teams. Good communication and problem-solving skills are important.
